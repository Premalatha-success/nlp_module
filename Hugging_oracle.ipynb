```python
### classification
from transformers import pipeline

# Load the sentiment-analysis pipeline
classifier = pipeline("text-classification", model="distilbert-base-uncased-finetuned-sst-2-english")

# Classify a sample text
text = "I love using Hugging Face! It's an amazing tool."
output = classifier(text)

print(output)
```

```python
from datasets import load_dataset

# Load a sample dataset (e.g., IMDb for sentiment classification)
dataset = load_dataset("imdb")
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")

def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True)

tokenized_dataset = dataset.map(tokenize_function, batched=True)
```

```python
from transformers import AutoModelForSequenceClassification

model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased", num_labels=2)
```

```python
from transformers import Trainer, TrainingArguments

training_args = TrainingArguments(output_dir="./results", evaluation_strategy="epoch")

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["test"]
)

trainer.train()
```

```python
from transformers import pipeline

# Load the NER pipeline
ner_pipeline = pipeline("ner", model="dbmdz/bert-large-cased-finetuned-conll03-english", grouped_entities=True)

# Example text
text = "Hugging Face Inc. is based in New York City, and its CEO is Clément Delangue."

# Perform NER
ner_results = ner_pipeline(text)

# Print the output
for entity in ner_results:
    print(entity)
```

```python
texts = [
    "Apple is looking at buying U.K. startup for $1 billion.",
    "Microsoft was founded by Bill Gates and Paul Allen in Albuquerque."
]

for text in texts:
    ner_results = ner_pipeline(text)
    print(f"\nEntities in '{text}':")
    for entity in ner_results:
        print(entity)
```

```python
from datasets import load_dataset

dataset = load_dataset("conll2003")
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")

def tokenize_and_align_labels(examples):
    tokenized_inputs = tokenizer(examples["tokens"], truncation=True, is_split_into_words=True)
    return tokenized_inputs

tokenized_dataset = dataset.map(tokenize_and_align_labels, batched=True)
```

```python
from transformers import Trainer, TrainingArguments

training_args = TrainingArguments(output_dir="./results", evaluation_strategy="epoch")

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["test"]
)

trainer.train()
```

```python
from transformers import pipeline

# Load the Q&A pipeline
qa_pipeline = pipeline("question-answering", model="distilbert-base-uncased-distilled-squad")

# Define the context and question
context = """
Hugging Face is a technology company that develops tools for building applications using machine learning.
It is based in New York City, with additional operations in Paris. Hugging Face is known for its open-source library
'transformers', which allows users to easily integrate state-of-the-art models like BERT, GPT, and more into their applications.
"""

question = "Where is Hugging Face located?"

# Get the answer
answer = qa_pipeline(question=question, context=context)

print(answer)
```

```python
questions = [
    "Where is Hugging Face located?",
    "What is Hugging Face known for?"
]

for question in questions:
    answer = qa_pipeline(question=question, context=context)
    print(f"Q: {question}\nA: {answer['answer']}\n")
```

```python
from datasets import load_dataset

squad = load_dataset("squad")
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

def preprocess_function(examples):
    return tokenizer(examples['question'], examples['context'], truncation=True, padding="max_length")

tokenized_dataset = squad.map(preprocess_function, batched=True)
```

```python
from transformers import AutoModelForQuestionAnswering

model = AutoModelForQuestionAnswering.from_pretrained("bert-base-uncased")
```

```python
from transformers import Trainer, TrainingArguments

training_args = TrainingArguments(
    output_dir="./results",
    evaluation_strategy="epoch"
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset["train"],
    eval_dataset=tokenized_dataset["validation"]
)

trainer.train()
```
